# 1. Read a color image from the Internet (i.e., nature.jpg). Display its red, green and blue component separately

```python
import cv2
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

url = 'https://minhhieu9800.github.io/mse_divp/nature.jpeg'

# read color image
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)
    img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# split into color channels
b, g, r = cv2.split(img)

# display each channel separately
plt.figure(figsize=(12, 12))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(cv2.cvtColor(r, cv2.COLOR_BGR2RGB))
plt.title('Red component')

plt.subplot(1, 4, 3)
plt.imshow(cv2.cvtColor(g, cv2.COLOR_BGR2RGB))
plt.title('Green component')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB))
plt.title('Blue component')

plt.show()
```

# 2. Swap the red and the blue component of the input image to create a new image (BGR), and save the new image into a new file

```python
import cv2
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# read input image
url = 'https://minhhieu9800.github.io/mse_divp/nature.jpeg'

# read color image
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)
    img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# swap red and blue channels
bgr_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# display image and histograms
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('RGB Image')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB))
plt.title('BGR Image')


plt.show()


# save new image
cv2.imwrite('output_image_ex2.jpg', bgr_img)
```

# 4. Display the image and its histogram in color and grayscale.

```python
import cv2
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# read input image
url = 'https://minhhieu9800.github.io/mse_divp/nature.jpeg'

# read color image
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)
    img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

# convert image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# calculate histograms for color and grayscale images
color_hist = cv2.calcHist([img], [0, 1, 2], None, [256, 256, 256], [0, 256, 0, 256, 0, 256])
gray_hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])

# display image and histograms
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Color Image')

plt.subplot(2, 2, 2)
plt.plot(color_hist.reshape(-1,))
plt.title('Color Histogram')

plt.subplot(2, 2, 3)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(2, 2, 4)
plt.plot(gray_hist.reshape(-1,))
plt.title('Grayscale Histogram')

plt.show()
```

# 5. With grayscale image, do three different actions: 
- flip left to right 
- rotate clockwise 180 degrees 
- crop  central of the image

```python
import cv2
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# read input image
url = 'https://minhhieu9800.github.io/mse_divp/nature.jpeg'

# read color image
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)
    img = cv2.imdecode(img_array, cv2.IMREAD_GRAYSCALE)

# flip image left to right
flipped_img = cv2.flip(img, 1)

# rotate image clockwise 180 degrees
rotated_img = cv2.rotate(img, cv2.ROTATE_180)

print(img.shape)

# crop image 1/2 central region
height, width = img.shape
crop_img = img[int(height/4):int(3*height/4), int(width/4):int(3*width/4)]

# display original and modified images
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(flipped_img, cmap='gray')
plt.title('Flipped Image')

plt.subplot(2, 2, 3)
plt.imshow(rotated_img, cmap='gray')
plt.title('Rotated Image')

plt.subplot(2, 2, 4)
plt.imshow(crop_img, cmap='gray')
plt.title('Cropped Image')

plt.show()


# save new images
cv2.imwrite('flipped_image_ex5.jpg', flipped_img)
cv2.imwrite('rotated_image_ex5.jpg', rotated_img)
cv2.imwrite('crop_image_ex5.jpg', crop_img)
```

# 6. Quantize the grayscale using 2 bits, 4 bits, 6 bits and 8 bits. Visualize the effect of the operations.

```python
import cv2
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# read input image
url = 'https://minhhieu9800.github.io/mse_divp/nature.jpeg'

# read grayscale image
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)
    img_gray = cv2.imdecode(img_array, cv2.IMREAD_GRAYSCALE)


# Define the number of bits for quantization
num_bits = [2, 4, 6, 8]

# Define the figure size and subplot layout
fig = plt.figure(figsize=(12, 4))
rows = 1
cols = len(num_bits) + 1

# Plot the original grayscale image in the first subplot
plt.subplot(rows, cols, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original')

# Loop over the number of bits and apply quantization
for i, n in enumerate(num_bits):
    # Calculate the maximum pixel value for this bit depth
    max_val = 2**n - 1
    
    # Quantize the image to the given bit depth
    img_quantized = np.floor_divide(img_gray, 256//max_val) * (256//max_val)
    
    # Convert the quantized image back to uint8
    img_quantized = img_quantized.astype(np.uint8)
    
    # Plot the quantized image in the corresponding subplot
    plt.subplot(rows, cols, i+2)
    plt.imshow(img_quantized, cmap='gray')
    plt.title(str(n) + ' bits')
    
# Show the figure
plt.tight_layout()
plt.show()
```

# 7. Perform histogram equalization (global and local) for grayscale and color images. Display the original and equalized ones.

## Grayscale image

```python
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# read input image
url = 'https://minhhieu9800.github.io/mse_divp/nature.jpeg'

# read grayscale image
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)
    img_gray = cv2.imdecode(img_array, cv2.IMREAD_GRAYSCALE)

# Global histogram equalization
img_eq_global = cv2.equalizeHist(img_gray)

# Local histogram equalization
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img_eq_local = clahe.apply(img_gray)

# Display original and equalized images
plt.figure(figsize=(12, 12))

plt.subplot(1, 3, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Gray Image')

plt.subplot(1, 3, 2)
plt.imshow(img_eq_global, cmap='gray')
plt.title('Global histogram equalization')

plt.subplot(1, 3, 3)
plt.imshow(img_eq_local, cmap='gray')
plt.title('Local histogram equalization')

plt.show()
```

## Color image

```python
import cv2
import numpy as np
import urllib.request
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

# read input image
url = 'https://minhhieu9800.github.io/mse_divp/nature.jpeg'

# read color image
with urllib.request.urlopen(url) as url_response:
    img_array = np.asarray(bytearray(url_response.read()), dtype=np.uint8)
    img_color = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

## Global
# Convert to HSV color space
img_hsv = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)

# Equalize value channel
img_hsv[:,:,2] = cv2.equalizeHist(img_hsv[:,:,2])

# Convert back to RGB color space
img_eq_global = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)


## Local
# Convert to LAB color space
img_lab = cv2.cvtColor(img_color, cv2.COLOR_BGR2LAB)

# Split LAB channels
l_channel, a_channel, b_channel = cv2.split(img_lab)

# Create CLAHE object with specified parameters
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Apply CLAHE to L channel
l_channel_equalized = clahe.apply(l_channel)

# Merge LAB channels back into color image
img_lab_equalized = cv2.merge((l_channel_equalized, a_channel, b_channel))

# Convert back to BGR color space
img_eq_local = cv2.cvtColor(img_lab_equalized, cv2.COLOR_LAB2BGR)



### Display original and equalized images
plt.figure(figsize=(12, 12))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title('original Image')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img_eq_global, cv2.COLOR_BGR2RGB))
plt.title('Global histogram equalization')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(img_eq_local, cv2.COLOR_BGR2RGB))
plt.title('Local histogram equalization')

plt.show()
```
