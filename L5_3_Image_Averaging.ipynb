L5_3_Image_Averaging

```python
import cv2
import numpy as np
import urllib.request
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt

# load image from URL
# img = io.imread('https://minhhieu9800.github.io/mse_divp/day_3/quadnight.jfif')

url = 'https://minhhieu9800.github.io/mse_divp/day_3/quadnight.jfif'
resp = urllib.request.urlopen(url)
img = np.asarray(bytearray(resp.read()), dtype='uint8')
img = cv2.imdecode(img, cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = img.astype(float) / 255.0

img = cv2.resize(img, (200, 200))


# cv2_imshow(img * 255)

# repeat over number of sample images
nArray = [1, 2, 8, 16, 32, 64]
for i in range(len(nArray)):
    N = nArray[i]
    avgImg = np.zeros(img.shape)  # initialize an empty image

    # add Gaussian noise to each image, accumulate this image to average image
    for j in range(N):
        noise = cv2.randn(np.zeros_like(img, dtype='float32'), 0, np.sqrt(0.03))
        noisyImg = img + noise
        avgImg += noisyImg

    # show image
    avgImg = avgImg / N
    print(str(N) + ' image(s)')
    cv2_imshow(avgImg * 255)
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```
