```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from skimage import io

# load test images
liveImg = io.imread('https://minhhieu9800.github.io/mse_divp/day_3/live.jpeg')
maskImg = io.imread('https://minhhieu9800.github.io/mse_divp/day_3/mask.jpeg')

# convert to float and scale to [0, 1]
# liveImg = np.double(liveImg) / 255.0
# maskImg = np.double(maskImg) / 255.0

# calculate the difference between two images
diffImg = np.abs(np.double(maskImg) / 255.0 - np.double(liveImg) / 255.0)
gamma = 0.3
diffImg_contrasted = np.power(diffImg, gamma)
```

Original Image

```python
cv2_imshow(liveImg)
```

Mask Image

```python
cv2_imshow(maskImg)
```

Difference Image

```python
cv2_imshow(diffImg * 255)
```

Difference Image increased contrast

```python
cv2_imshow(diffImg_contrasted * 255)
```

```python
cv2.waitKey(0)
```

```python

# save images
cv2.imwrite('Image_Subtraction.jpeg', diffImg * 255)
cv2.imwrite('Image_Subtraction_with_constrast.jpeg', diffImg_contrasted * 255)
cv2.destroyAllWindows()
```
